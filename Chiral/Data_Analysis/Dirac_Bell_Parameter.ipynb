```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
df_Bell_Parameter_Ress_O = pd.read_csv(r'Bell_Parameter/Bell_Ress/Dirac_Bell_Parameter_O_Ress.txt', sep = '\t', header = None)
df_Bell_Parameter_Ress_U = pd.read_csv(r'Bell_Parameter/Bell_Ress/Dirac_Bell_Parameter_U_Ress.txt', sep = '\t', header = None)
```

```python
df_Bell_Parameter_Ress_O.rename(columns = lambda s: ((s*25)+50))
```

```python
df_Bell_Parameter_Ress_U.rename(columns = lambda s: ((s*25)+50))
```

```python
Bell_Parameter_Ress_O = df_Bell_Parameter_Ress_O.values
Bell_Parameter_Ress_U = df_Bell_Parameter_Ress_U.values

print("VIOLATIONS OF BELL PARAMETER IN UNITARY ENSEMBLE")
print(df_Bell_Parameter_Ress_U[df_Bell_Parameter_Ress_U > 2.0].count())
print("\n ########################################## \n")
print("VIOLATIONS OF BELL PARAMETER IN ORTHOGONAL ENSEMBLE")
print(df_Bell_Parameter_Ress_O[df_Bell_Parameter_Ress_O > 2.0].count())
```

```python
Violations_U = df_Bell_Parameter_Ress_U[df_Bell_Parameter_Ress_U > 2.0].count().values
Violations_O = df_Bell_Parameter_Ress_O[df_Bell_Parameter_Ress_O > 2.0].count().values
print("MAX VALUES OF BELL PARAMETER IN UNITARY ENSEMBLE")
print(df_Bell_Parameter_Ress_U.max())
print("\n ######################################### \n")
print("MAX VALUES OF BELL PARAMETER IN ORTHOGONAL ENSEMBLE")
print(df_Bell_Parameter_Ress_O.max())
```

```python
fig, axs = plt.subplots(1,2, figsize = (14, 6), constrained_layout = True)

fig.suptitle(r'HISTOGRAM OF $P(|\mathcal{S}_{CHSH}|)$ FOR SCHRÖDINGER BILLIARD WITH 300 RESONANCES', fontsize = 20)

axs[0].hist(Bell_Parameter_Ress_O[:,10], bins = 50, color = 'dodgerblue', label = r'TRS', alpha = 1, density = True, histtype = 'bar', ec = 'black')
axs[0].legend(prop = {'size': 20}, bbox_to_anchor = (0.7, 0.97))
axs[0].set_xlabel(r'$|\mathcal{S}_{CHSH}|$', fontsize = 20)
axs[0].tick_params(axis = 'both', which = 'major', labelsize = 20)
axs[0].set_ylabel('$P(|\mathcal{S}_{CHSH}|)$', fontsize = 20)
axs[0].axvline(2, color = 'black')

counts_U, bins_U = np.histogram(Bell_Parameter_Ress_U, bins = 50)
axs[1].hist(Bell_Parameter_Ress_U[:,10], bins = 50, color = 'red', label = r'No TRS', alpha = 1, density = True, histtype = 'bar', ec = 'black')
axs[1].legend(prop = {'size': 20}, bbox_to_anchor = (0.7, 0.97))
axs[1].set_xlabel(r'$|\mathcal{S}_{CHSH}|$', fontsize = 20)
axs[1].tick_params(axis = 'both', which = 'major', labelsize = 20)
axs[1].set_ylabel('$P(|\mathcal{S}_{CHSH}|)$', fontsize = 20)
axs[1].axvline(2, color = 'black')
```

```python
fig, axs = plt.subplots(1,1, figsize = (8, 6), constrained_layout = True)

axs.scatter([25*i + 50 for i in range(len(Violations_U))], (100/1000000)*Violations_O, label = 'TRS', s = 100, color = 'dodgerblue', alpha = 1)
axs.scatter([25*i + 50 for i in range(len(Violations_U))], (100/1000000)*Violations_U, label = 'No TRS', s = 100, color = 'red', alpha = 1)
    
axs.plot([25*i + 50 for i in range(0,11)], (100/1000000)*Violations_O, color = 'dodgerblue', linestyle = 'dashed')
axs.plot([25*i + 50 for i in range(0,11)], (100/1000000)*Violations_U, color = 'red', linestyle = 'dashed')

axs.set_ylim((0,2.5))

fig.suptitle(r'VIOLATIONS OF BELL INEQUALITY IN TERMS OF RESONANCES', fontsize = 20)
axs.set_xlabel('Resonances', fontsize = 15)
axs.set_ylabel('Violations (%)', fontsize = 19)
axs.tick_params(axis = 'both', which = 'major', labelsize = 15)
axs.grid(linestyle = ':')
axs.legend(fontsize = 15)
```

```python
df_Bell_Parameter_Gamma_O = pd.read_csv(r'Bell_Parameter/Bell_Gamma/Dirac_Bell_Parameter_O_Gamma.txt', sep = '\t', header = None)
df_Bell_Parameter_Gamma_U = pd.read_csv(r'Bell_Parameter/Bell_Gamma/Dirac_Bell_Parameter_U_Gamma.txt', sep = '\t', header = None)
```

```python
df_Bell_Parameter_Gamma_O.rename(columns = lambda s: (s/20))
```

```python
df_Bell_Parameter_Gamma_U.rename(columns = lambda s: (s/20))
```

```python
print("MAX VALUES OF BELL PARAMETER IN ORTHOGONAL ENSEMBLE")
print(df_Bell_Parameter_Gamma_O.max())
print("\n")
print("MAX VALUES OF BELL PARAMETER IN UNITARY ENSEMBLE")
print(df_Bell_Parameter_Gamma_U.max())
```

```python
Violations_Gamma_O = df_Bell_Parameter_Gamma_O[df_Bell_Parameter_Gamma_O > 2.0].count().values
print("VIOLATIONS OF BELL'S INEQUALITY IN ORTHOGONAL ENSEMBLE")
print(df_Bell_Parameter_Gamma_O[df_Bell_Parameter_Gamma_O > 2.0].count())
print("\n")
print("VIOLATIONS OF BELL'S INEQUALITY IN UNITARY ENSEMBLE")
Violations_Gamma_U = df_Bell_Parameter_Gamma_U[df_Bell_Parameter_Gamma_U > 2.0].count().values
print(df_Bell_Parameter_Gamma_U[df_Bell_Parameter_Gamma_U > 2.0].count())
```

```python
fig, axs = plt.subplots(1,1, figsize = (8, 6), constrained_layout = True)

axs.scatter([(i/20) for i in range(0,21)], (100/1000000)*Violations_Gamma_O, label = 'TRS', s = 100, color = 'dodgerblue', alpha = 1)
axs.scatter([(i/20) for i in range(0,21)], (100/1000000)*Violations_Gamma_U, label = 'No TRS', s = 100, color = 'red', alpha = 1)
    
axs.plot([(i/20) for i in range(0,21)], (100/1000000)*Violations_Gamma_O, color = 'dodgerblue', linestyle = 'dashed')
axs.plot([(i/20) for i in range(0,21)], (100/1000000)*Violations_Gamma_U, color = 'red', linestyle = 'dashed')

axs.set_ylim((0,4))

fig.suptitle(r'VIOLATIONS OF BELL INEQUALITY AS FUNCTION OF $\Gamma$', fontsize = 20)
axs.set_xlabel(r'$\Gamma$', fontsize = 15)
axs.set_ylabel('Violations (%)', fontsize = 19)
axs.tick_params(axis = 'both', which = 'major', labelsize = 15)
axs.grid(linestyle = ':')
axs.legend(fontsize = 15)
```
